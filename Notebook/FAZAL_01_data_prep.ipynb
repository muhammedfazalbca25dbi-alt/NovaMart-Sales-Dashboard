# 2.De-duplicate

```python
nova_all.drop_duplicates()
```

# 3.Fix the sales column

```python
nova_all["sales"] = nova_all["sales"].astype(str).str.replace("$", "", regex=False)
nova_all["sales"] = nova_all["sales"].str.replace(",", "", regex=False)
nova_all["sales"] = nova_all["sales"].astype(float)
```

```python
nova_all['sales'].head()
```

# 4.Parse dates

```python
nova_all["order_date"] = pd.to_datetime(nova_all["order_date"], format='mixed')
nova_all["ship_date"] = pd.to_datetime(nova_all["ship_date"], format='mixed')
```

# 5. Standardise text

```python
nova_all = nova_all.rename(columns={
    "region_y": "region",
    "category_y": "category",
    "sub_category_y": "sub_category"
})
```

```python
nova_all["category"] = nova_all["category"].str.strip().str.title()
```

```python
nova_all["region"] = nova_all["region"].str.strip().str.title()
```

# 6. Handle missing values

```python
nova_all["discount"] = nova_all["discount"].fillna(0)
```

```python
nova_all["profit"] = nova_all["profit"].fillna(0)
```

# 7.Engineer columns: order_month, profit_margin_pct, shipping_days

```python
nova_all["order_month"] = nova_all["order_date"].dt.month_name()

nova_all["shipping_days"] = (nova_all["ship_date"] - nova_all["order_date"]).dt.days

nova_all["profit_margin_pct"] = (nova_all["profit"] / nova_all["sales"]) * 100
```

# 8.Enrich

```python
product = pd.read_csv(r"C:\Users\muham\OneDrive\Documents\Project2_NovaMart_Streamlit\Project2_NovaMart_Streamlit\data\product_lookup.csv ")

nova_all = nova_all.merge(product, on="product_id", how="left")
```

```python
customer = pd.read_csv(r"C:\Users\muham\OneDrive\Documents\Project2_NovaMart_Streamlit\Project2_NovaMart_Streamlit\data\customer_lookup.csv")

nova_all = nova_all.merge(customer, on="customer_id", how="left")
```

```python
nova_all = nova_all.drop(columns=[
    "region_x",
    "category_x",
    "sub_category_x"
])
```
